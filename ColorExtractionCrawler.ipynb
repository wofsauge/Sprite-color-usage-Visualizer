```python
import glob
from PIL import Image
from collections import defaultdict
```

```python
path = "D:\\Programme/Steam/steamapps/common/The Binding of Isaac Rebirth/resources/gfx/monsters"
outputFileName = 'data/BoI_monsters_colors.csv'
```

```python
# itterate over every file in every subfolder and get some color data
files = glob.glob(path + '/**/*.png', recursive=True)
by_color = defaultdict(int)
colorsUsedByNumFiles = defaultdict(int)
for file in files:
    im = Image.open(file)
    colorsUsedByThisFile = defaultdict(int)
    for pixel in im.getdata():
        if pixel[0]+pixel[1]+pixel[2]>0:
            colorString = str(pixel[0])+","+str(pixel[1])+","+str(pixel[2])+","+str(pixel[3])
            by_color[colorString] += 1
            colorsUsedByThisFile[colorString] += 1
            
    for color in colorsUsedByThisFile.keys():
        colorsUsedByNumFiles[color] += 1
        
    print("done "+file.replace(path, ""))
```

```python
# Write color datas into a nice csv file
import csv
with open(outputFileName, 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=';', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(["R","G","B","A","Count","FileCount"])
    for color in by_color.keys():
        split = color.split(',')
        spamwriter.writerow([split[0],split[1],split[2],split[3],by_color[color],colorsUsedByNumFiles[color]])
```
