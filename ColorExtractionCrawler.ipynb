```python
import glob
from PIL import Image
from collections import defaultdict
```

```python
path = "D:\\Programme/Steam/steamapps/common/The Binding of Isaac Rebirth/resources/gfx/items"
files = glob.glob(path + '/**/*.png', recursive=True)
```

```python
by_color = defaultdict(int)
for file in files:
    im = Image.open(file)
    for pixel in im.getdata():
        if pixel[0]+pixel[1]+pixel[2]>0:
            by_color[str(pixel[0])+","+str(pixel[1])+","+str(pixel[2])+","+str(pixel[3])] += 1
    print("done "+file)
```

```python
import csv
with open('items_colors.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=';', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(["R","G","B","A","Count"])
    for color in by_color.keys():
        split = color.split(',')
        spamwriter.writerow([split[0],split[1],split[2],split[3],by_color[color]])
```
