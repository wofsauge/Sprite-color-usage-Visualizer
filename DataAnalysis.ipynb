```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

```python
filename = "BoI_monsters_colors.csv"
```

```python

# Helix equation
data = pd.read_csv("data/"+filename,sep=";") 
data = data.drop(data[(data.A == 0)].index)
data = data.drop(data[(data.R == 8) &(data.G == 0) &(data.B == 0)].index)
data = data.drop(data[(data.R == 255) &(data.G == 255) &(data.B == 255)].index)

# build color lookup table
colors = []
for k,v in pd.DataFrame(data).iterrows():
    #colors.append('rgba('+str(v[0])+','+str(v[1])+','+str(v[2])+','+str(v[3])+')')
    colors.append('#%02x%02x%02x' % (v[0], v[1], v[2]))

fig = go.Figure(data=[go.Scatter3d(
    x=data['R'],
    y=data['G'],
    z=data['B'],
    text=data['Count'],
    mode='markers',
    marker=dict(
        sizemin=5,
        sizemode='area',
        sizeref=2.*max(data['Count'])/50/(25.**2),
        size=data['Count'],
        color=colors,
        opacity=0.85,
        line_color=colors
    ),
    hovertemplate =
    '<b>Red</b>: %{x}'+
    '<br><b>Green</b>: %{y}'+
    '<br><b>Blue</b>: %{z}'+
    '<br><b>Usages</b>: %{text}<extra></extra>',
)])
fig.update_layout(width=1300, height=800, title = 'Colors',titlefont_color='white',
                  scene = dict(xaxis=dict(title="R",titlefont_color='white', showbackground=False),
                               yaxis=dict(title="G",titlefont_color='white', showbackground=False),
                               zaxis=dict(title="B",titlefont_color='white', showbackground=False),
                               bgcolor = 'rgb(20, 24, 54)'
                           ))
# tight layout
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0));
```

```python
fig.show()
```

```python
fig.write_html("results/"+filename+".html")
```
